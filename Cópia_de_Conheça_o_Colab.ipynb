```
meses = {
    1: 'Janeiro', 2: 'Fevereiro', 3: 'Março', 4: 'Abril',
    5: 'Maio', 6: 'Junho', 7: 'Julho', 8: 'Agosto',
    9: 'Setembro', 10: 'Outubro', 11: 'Novembro', 12: 'Dezembro'
}

exp['CO_MES'] = exp['CO_MES'].replace(meses)
```

```
# ADICIONANDO UMA NOVA COLUNA VALOR AGREGADO
# Garante que VL_FOB e KG_LIQUIDO sejam numéricos
```

```
exp['VL_FOB'] = pd.to_numeric(exp['VL_FOB'], errors='coerce')
exp['KG_LIQUIDO'] = pd.to_numeric(exp['KG_LIQUIDO'], errors='coerce')
```

```
#(valores 0, nulos ou infinitos) Removendo linhas inválidas
exp = exp[(exp['VL_FOB'] > 0) & (exp['KG_LIQUIDO'] > 0)]
```

```
# Calcula corretamente o valor agregado
exp['valor_agregado'] = (exp['VL_FOB'] / exp['KG_LIQUIDO']).round(2)
```

```
# Substitui infinitos e nulos
exp['valor_agregado'].replace([np.inf, -np.inf], np.nan, inplace=True)
exp['valor_agregado'].fillna(0, inplace=True)
```

```
# Formata o número no padrão brasileiro (opcional)
exp['valor_agregado'] = exp['valor_agregado'].apply(
    lambda x: f"{x:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
```

```
print(exp['valor_agregado'].head())
```

```
#HHI por ano (por país)
#Soma por (ano,país)
```

```
exp_ano_pais = exp.groupby(['CO_ANO', 'CO_PAIS'], dropna=False)['VL_FOB'].sum().reset_index(name='vl_fob_ano_pais')
```

```
#total por ano
total_ano = exp_ano_pais.groupby('CO_ANO')['vl_fob_ano_pais'].sum().reset_index(name='total_ano')
```

```
# juntar e calcular participação
exp_ano_pais = exp_ano_pais.merge(total_ano, on='CO_ANO', how='left')
exp_ano_pais['participacao'] = exp_ano_pais['vl_fob_ano_pais'] / exp_ano_pais['total_ano']
```

```
# lidar com possíveis NaN/inf em participacao
exp_ano_pais['participacao'].replace([np.inf, -np.inf], np.nan, inplace=True)
exp_ano_pais['participacao'].fillna(0, inplace=True)
```

```
# HHI (soma dos quadrados das participações) por ano
hhi = (
    exp_ano_pais.groupby('CO_ANO')['participacao']
    .apply(lambda s: (s**2).sum())
    .reset_index(name='HHI'))
```

```
#  multiplicar por 10000 para escala clássica (0-10000)
hhi['HHI_10000'] = (hhi['HHI'] * 10000).round(2)

print("HHI (amostra):")
print(hhi.head())
```
